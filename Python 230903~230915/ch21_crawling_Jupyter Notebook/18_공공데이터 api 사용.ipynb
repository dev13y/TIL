#### 국립 중앙의료원 제공 전국 약국 정보 데이터  
- 요청변수가 코드가 아닌 실제 지역명 등 한글 이름 사용해야 함  
- 안글 인코딩 처리 : quote()

```python
import pandas as pd
from urllib.parse import quote
import requests
import bs4
```

```python
# 요청 주소
endpoint = 'http://apis.data.go.kr/B552657/ErmctInsttInfoInqireService/getParmacyListInfoInqire?'
```

```python
# 본인의 serviceKey 입력
serviceKey = '#'
```

```python
# 파라미터 변수 설정
QT = '1' # 월요일에 여는 약국
QN = quote('삼성약국') # 기관명 (약국이름) 인코딩 -> 삼성약국'이 포함된 모든 약국 반환
Q0 = quote('서울특별시') # 광역시도
Q1 = quote('강남구')  # 시군구
ORD = 'NAME' # 정렬기준 : 기관명
pageNo = '1'
numOfRows = '10' # 목록 건수
```

```python
# 파라미터 설정
params = "serviceKey=" + serviceKey \
        + "&Q0=" + Q0 \
        + "&Q1=" + Q1 \
        + "&QN=" + QN \
        + "&QT=" + QT \
        + "&ORD=" + ORD \
        + "&pageNo=" + pageNo \
        + "&numOfRows=" + numOfRows 
```

```python
# 요청 url
url = endpoint + params
```

```python
url
```

```python
# 서비스 요청 및 결과 파싱
result = requests.get(url)
result
bs_obj = bs4.BeautifulSoup(result.content, 'html.parser')
bs_obj
```

```python
# item 추출
items = bs_obj.findAll('item')
len(items)
```

```python
# 데이터 추출
# 반환된 결과에서 약국이름 <dutyname>, 주소 추출 <dutyaddr>
for item in items:
    name = item.find('dutyname').text
    address = item.find('dutyaddr').text
    print(name, address)
```

#### 서울의 모든 약국

```python
# 서울의 모든 약국 : 광역시도 = 서울특별시
# Q0만 사용
```

```python
# 파라미터 변수 설정
QT = '1' # 월요일에 여는 약국
QN = quote('삼성약국') # 기관명 (약국이름) 인코딩 -> 삼성약국'이 포함된 모든 약국 반환
Q0 = quote('서울특별시') # 광역시도
Q1 = quote('강남구')  # 시군구
ORD = 'NAME' # 정렬기준 : 기관명
pageNo = '1'
numOfRows = '50' # 목록 건수
```

```python
# 파라미터 설정
params = "serviceKey=" + serviceKey \
        + "&Q0=" + Q0 \
        + "&ORD=" + ORD \
        + "&numOfRows=" + numOfRows 
```

```python
# 요청 url
url = endpoint + params
```

```python
# 서비스 요청 및 결과 파싱
result = requests.get(url)
result
bs_obj = bs4.BeautifulSoup(result.content, 'html.parser')
bs_obj
```

```python
items = bs_obj.findAll('item')
len(items)
```

```python
# 서울시 약국 정보(약국명, 전화, 주소 추출해서) 데이터프레임으로 저장
# items = bs_obj.findAll('items')
# items

name = []
phone = []
addr = []


for item in items:
    name.append(item.find('dutyname'))
    phone.append(item.find('dutytel1'))
    addr.append(item.find('dutyaddr'))


data = {'약국이름': name, '전화번호': phone, '주소': addr}
drugs_df = pd.DataFrame(data)
drugs_df
```

```python
items = bs_obj.select('item')

name = [item.select_one('dutyname').text for item in items]
tel = [item.select_one('dutytel1').text for item in items]
addr = [item.select_one('dutyaddr').text for item in items]

df = pd.DataFrame({
    '약국명':name,
    '전화':tel,
    '주소':addr
})
df
```

```python
# 월요일 밤 9시 이후까지 영업하는 약국 추출 : 2100
# 월요일 영업 <dutytime1c>
# 약국 수 출력 
# 데이터프레임 저장
```

```python
items = bs_obj.findAll('items')

name = []
phone = []
addr = []
m_time = []

items = bs_obj.findAll('items')[0]

for item in items:
    if int(item.find('dutytime1c').text.strip()) >= 2100:
        name.append(item.find('dutyname'))
        phone.append(item.find('dutytel1'))
        addr.append(item.find('dutyaddr'))
        m_time.append(item.find('dutytime1c'))
    
data = {'약국이름': name, '전화번호': phone, '주소': addr, '월요일': m_time}
drugs_df = pd.DataFrame(data)
drugs_df
```

```python
# 2
dutyaddr, dutyname, dutytel = [[] for _ in range(3)]

for item in bs_obj.findAll('item'):
    if int(item.find('dutytime1c').text) >= 2100:
        dutyname.append(item.find('dutyname').text)
        dutytel.append(item.find('dutytel1').text)
        dutyaddr.append(item.find('dutyaddr').text)

duty_df = pd.DataFrame({'약국명':dutyname, '전화':dutytel, '주소':dutyaddr})
duty_df
```
