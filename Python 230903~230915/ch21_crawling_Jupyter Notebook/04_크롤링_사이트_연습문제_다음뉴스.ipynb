- 'https://news.daum.net' 에서 카드뉴스의 섹션과 제목을 추출해서
- 데이터프레임으로 생성 후 daum_news_section.csv로 저장

```python
from urllib.request import urlopen
import bs4
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
import pandas as pd
```

```python
# 사이트 연결. HTML 파싱
url = 'https://news.daum.net'
html = urlopen(url)
bs_obj = bs4.BeautifulSoup(html, 'html.parser')

lis = bs_obj.select('.list_newsissue li')

sections = []
titles = []

for li in lis:
    section = li.select_one('.txt_category').text
    title = li.select_one('.tit_g a').text.strip()
    sections.append(section)
    titles.append(title)
    
news_section_df = pd.DataFrame({'section':sections, 'title':titles})
news_section_df.to_csv('./crawl_data/daum_news_section.csv')
```
