#### 검색 기능을 함수로 구현  
- 결과 최대 100개   
- 100개 이상 수집하려면 페이지(start) 이동하면서 여러 번 호출해야 함  
- 함수로 구현해 놓고 여러 번 호출해서 사용  
- get_api_result(keyword, display)  

```python
# keyword(검색어)와 display(검색 결과 수) 전달받아서
# API 호출 결과를 반환하는 함수
def get_api_result(keyword, display):
    # id, scret
    # client_id = ""
    # client_secret = ""

    # 키워드 
    keyword = urllib.parse.quote(keyword)

    # 요청 url + 파라미터
    base_url = "https://openapi.naver.com/v1/search/blog.json?"
    param = "query=" + keyword + "&display=" + str(display)
    url = base_url + param

    # headers 인증벙보
    headers = {
    "X-Naver-Client-Id":client_id,
    "X-Naver-Client-Secret":client_secret
    }

    # 요청 및 결과 반환
    result = requests.get(url, headers=headers)

    # json 변환
    json_obj = result.json()

    return json_obj
    
```

#### 여러 페이지 호출하는 함수 구현
- start 값에 따라 페이지 별로 쉽게 출력하기 위해  
- 출력 부분을 함수로 묶기  
- call_and_print(keyword, display, start):  
    - 검색어, 추출 개수, 시작 페이지 전달받아서  
    - 검색 결과에 대해 블로그 제목, 글링크,  블로거 이름을 출력하는 함수  
    - 함수 내에서 get_api_result(keyword, display, start) 함수 호출하고  
    - 결과 받아서 추출  

```python
def call_and_print(keyword, display, start):
    json_obj = get_api_result(keyword, display, start)

    for i, item in enumerate(json_obj['items']):
        print(i+1, ".", item['title'].replace('<b>','').replace('</b>',''), item['link'])
        
```

#### 뉴스/블로그 둘 다  검색 가능하게 함수 수정
- 블로그 또는 뉴스 검색 후 결과를 저장하는 함수로 구현
- 블로그와 뉴스에 대한 url 변경되게 구성
    - category 사용  
- title, link 추출  
- get_api_result() 함수 수정  
- call_and_save() 함수 수정  
- if cagtegory == '뉴스' :  
    - ctg = 'news'  
  else:  
    - ctg = 'blog'  
  base_url 변경  

- 원하는 검색 개수를 몇 번 호출할지 계산  
  - 302  
        - 100으로 나눈 몫 : 3 (100개)  
        - 나머지 : 2 (낱개)
        - display가 100으로 3번, 2로 1번 전달

```python
# # get_api_result() 수정 : category 추가
# # get_api_result(category, keyword, display, start):
def get_api_result(category, keyword, display, start):
    # id, scret
    # client_id = ""
    # client_secret = ""

    # 키워드 
    keyword = urllib.parse.quote(keyword)

    # category
    if category == '뉴스':
        ctg = "news"
    else:
        ctg = 'blog'

    # requests url + param
    base_url = "https://openapi.naver.com/v1/search/" + ctg + ".json?"
    
    param = "query=" + keyword + \
            "&display=" + str(display) + \
            "&start=" + str(start)
    
    url = base_url + param


    # 요청 및 결과 반환
    result = requests.get(url, headers=headers)

    # json 변환
    json_obj = result.json()

    return json_obj
```

```python
# # call_and_save() 수정
# # 요청하는 display와 start 계산해서 반복
# # 데이터프레임데 다 저장해서 데이터프레임 반환

def call_and_save(category, keyword, display):
    # 빈 데이터프레임 생성
    result_df = pd.DataFrame()
    
    # display, start 계산 : 몫과 나머지 구해서 display, star 값 설정
    # 반복문 302 : display가 100으로 3번, 2로 1번 전달
    quo = display // 100 # 나머지로 나눈 몫 : quotient
    rem = display % 100 # 나머지

    for i in range(quo + 1):
        if i == quo:
            display = rem
        else:
            display = 100
    
        json_obj = get_api_result(category, keyword, display, quo*100+1)
        # print(start)
    
        # 결과를 바로 리스트에 저장
        title = [item['title'].replace('<b>','').replace('</b>','') for item in json_obj['items']]
        link = [item['link'] for item in json_obj['items']]
    
        # 반복문 안 에서 데이터프레임에 추가
        df = pd.DataFrame({
            "제목":title,
            "링크":link
        })

        result_df = pd.concat([result_df, df], axis=0, ignore_index=True)

    return result_df
```

```python
# (2)
def get_api_result(category, keyword, display, start):
    #id, secret
    client_id = ""
    client_secret = ""
    
    # 키워드
    keyword = urllib.parse.quote(keyword)
    if category == '블로그':
        ctg = 'blog'
    elif category == '뉴스':
        ctg = 'news'
    elif category == '책':
        ctg = 'book'
    # 요청 url + 파라미터
    base_url = "https://openapi.naver.com/v1/search/"
    param = ctg + ".json?" + \
            'query=' + keyword + \
            '&display=' + str(display) + \
            '&start=' + str(start)

    # 최종 url
    url = base_url + param
    
    # headers 인증번호
    headers = {
        "X-Naver-Client-Id":client_id,
        "X-Naver-Client-Secret":client_secret
    }

    # 요청 및 결과 반환
    result = requests.get(url, headers=headers)
    
    # json 변환
    json_obj = result.json()
    
    return json_obj
```

```python
# (2) 
def call_and_save(category, keyword, display):
    start = 1
    df_final = pd.DataFrame()
    
    if display%100==0:
        r = display//100
    else:
        r = display//100+1
        
    for i in range(r):
        if display>100:
            # json_obj = get_api_result(category, keyword, 100, start)
            display -= 100
            start += 100
        else:
            json_obj = get_api_result(category, keyword, display, start)
        df = pd.DataFrame({
                    '제목' : [item['title'].replace('<b>', '').replace('</b>', '') for item in json_obj['items']],
                    '링크' : [item['link'] for item in json_obj['items']]
                })
        df_final = pd.concat([df_final, df], ignore_index=True)
    return df_final

```
